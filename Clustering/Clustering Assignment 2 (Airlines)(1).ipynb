```python
import numpy as np
import pandas as pd 

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import StandardScaler,MinMaxScaler,RobustScaler
```

```python
dataset = pd.read_excel("EastWestAirlines.xlsx", sheet_name='data')
```

```python
dataset.head()
```

```python
dataset = dataset.rename(columns={'ID#':'ID','Award?':'Award'})
dataset = dataset.drop(['ID','Award'],axis=1)
dataset.head()
```

```python
dataset.isnull()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
scaler = MinMaxScaler()
scaled_features=scaler.fit_transform(dataset)
```

```python
dataset
```

```python
scaled_features
```

# KMEANS

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
x = model.fit_predict(scaled_features)
```

```python
x
```

```python
model.cluster_centers_
```

```python
dataset1 = dataset.copy()
dataset1['clusterID']= x
dataset1
```

```python
plt.figure(figsize = (10,7))
plt.scatter(dataset1['clusterID'],dataset['Balance'],c = x)
```

# HIERARICAL 

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters = 5,affinity = 'l2',linkage = 'complete')
y = model.fit_predict(scaled_features)
y
```

```python
dataset2 = dataset.copy()
dataset2['clusteriD']= y
dataset2
```

```python
plt.figure(figsize=(10,7))
plt.scatter(dataset2['clusteriD'],dataset2['Balance'],c = y)
```

```python
from sklearn.cluster import DBSCAN
```

# DBSCAN

```python
model = DBSCAN(eps = 1, min_samples = 4)
model.fit(scaled_features)
```

```python
model.labels_
```

```python
dataset['cluster'] = model.labels_
dataset
```

```python
plt.figure(figsize=(10,7))
plt.scatter(dataset['cluster'],dataset['Balance'],c=model.labels_)
```

# ELBOW METHOD

```python
nc=range(1,11)
l=[]
for i in nc:

    model=KMeans(n_clusters=i)
    model.fit(scaled_features)
    l.append(model.inertia_)
```

```python
plt.figure(figsize=(10,4))
plt.style.use("seaborn-darkgrid")
plt.scatter(nc,l,s=100,color="black")
plt.plot(nc,l,linestyle="-.")
plt.xticks(nc)
plt.show()
```

# SILHOUETTE

```python
from sklearn.metrics import silhouette_score
```

```python
nc=range(2,11)
l=[]
for i in nc:

    model=AgglomerativeClustering(n_clusters=i,affinity='l2',linkage="complete")
    #model.fit_predict(scaled_fearures)
    l.append(silhouette_score(scaled_features,model.fit_predict(scaled_features)))
```

```python
plt.figure(figsize=(10,4))
plt.figure(figsize=(10,4))
plt.style.use("seaborn-darkgrid")
plt.scatter(nc,l,s=100,color="black")
plt.plot(nc,l,linestyle="-.")
plt.xticks(nc)
plt.show()
```
